```python
'''SOURCE CODE'''
import random
r=random.randrange(1,100)
guess="no"
while guess =="no":
    n=int(input("Enter number:"))
    if(n>r):
        print("Random number is smaller than guessed!")
        guess="no"
    elif(n<r):
        print("Random number is greater than guessed!")
        guess="no"
    else:
        print("Correct")
        guess="yes"
print("Done")
```
